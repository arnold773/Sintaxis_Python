```python
def ingresar_calificaciones():
    materias = []
    calificaciones = []

    while True:
        # Solicitar nombre de la materia
        materia = input("Ingrese el nombre de la materia: ")
        materias.append(materia)

        # Solicitar calificación con validación
        while True:
            try:
                calificacion = float(input(f"Ingrese la calificación de {materia} (0 a 10): "))
                if 0 <= calificacion <= 10:
                    calificaciones.append(calificacion)
                    break
                else:
                    print("⚠️ La calificación debe estar entre 0 y 10.")
            except ValueError:
                print("⚠️ Debe ingresar un número válido.")

        # Preguntar si desea continuar
        continuar = input("¿Desea ingresar otra materia? (s/n): ").lower()
        if continuar != "s":
            break

    return materias, calificaciones


# Ejemplo de uso
if __name__ == "__main__":
    materias, calificaciones = ingresar_calificaciones()
    print("\n📊 Materias y calificaciones registradas:")
    for m, c in zip(materias, calificaciones):
        print(f"{m}: {c}")
```

```python
def ingresar_calificaciones():
   
    calificaciones = []

    while True:
        

        # Solicitar calificación con validación
        while True:
            try:
                calificacion = float(input(f"Ingrese la calificación entre (0 a 10): "))
                if 0 <= calificacion <= 10:
                    calificaciones.append(calificacion)
                    break
                else:
                    print("⚠️ La calificación debe estar entre 0 y 10.")
            except ValueError:
                print("⚠️ Debe ingresar un número válido.")

        # Preguntar si desea continuar
        continuar = input("¿Desea ingresar otra calificacion? (s/n): ").lower()
        if continuar != "s":
            break

    return  calificaciones

def calcular_promedio(calificaciones):
    if len(calificaciones) == 0:
        return 0  # Evita división por cero si la lista está vacía
    return sum(calificaciones) / len(calificaciones)


# Ejemplo de uso
if __name__ == "__main__":
    calificacions = ingresar_calificaciones()
    promedio = calcular_promedio(calificacions)
    print(f"📊 El promedio de las calificaciones es: {promedio:.2f}")
```

```python
def ingresar_calificaciones():
   
    calificaciones = []

    while True:
        

        # Solicitar calificación con validación
        while True:
            try:
                calificacion = float(input(f"Ingrese la calificación entre (0 a 10): "))
                if 0 <= calificacion <= 10:
                    calificaciones.append(calificacion)
                    break
                else:
                    print("⚠️ La calificación debe estar entre 0 y 10.")
            except ValueError:
                print("⚠️ Debe ingresar un número válido.")

        # Preguntar si desea continuar
        continuar = input("¿Desea ingresar otra calificacion? (s/n): ").lower()
        if continuar != "s":
            break

    return  calificaciones

def determinar_estado(calificaciones, umbral=5.0):
    aprobadas = []
    reprobadas = []

    for i, nota in enumerate(calificaciones):
        if nota >= umbral:
            aprobadas.append(i)
        else:
            reprobadas.append(i)

    return aprobadas, reprobadas


# Ejemplo de uso
if __name__ == "__main__":
    calificacions = ingresar_calificaciones()
    aprobadas, reprobadas = determinar_estado(calificacions)

    print("✅ Índices de materias aprobadas:", aprobadas)
    print("❌ Índices de materias reprobadas:", reprobadas)
```

```python
def encontrar_extremos(calificaciones):
    if len(calificaciones) == 0:
        return None, None  # Evita errores si la lista está vacía

    indice_max = calificaciones.index(max(calificaciones))
    indice_min = calificaciones.index(min(calificaciones))

    return indice_max, indice_min


# Ejemplo de uso
if __name__ == "__main__":
    calificaciones = [8.5, 4.0, 6.2, 3.5, 9.0]
    indice_max, indice_min = encontrar_extremos(calificaciones)

    print(f"📈 Índice de la calificación más alta: {indice_max}")
    print(f"📉 Índice de la calificación más baja: {indice_min}")
```
