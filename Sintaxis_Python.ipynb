```python
def ingresar_calificaciones():
    materias = []
    calificaciones = []

    while True:
        materia = input("Ingrese el nombre de la materia: ")
        materias.append(materia)

        # Validar que la calificación sea un número entre 0 y 10
        while True:
            try:
                calificacion = float(input(f"Ingrese la calificación de {materia} (0 a 10): "))
                if 0 <= calificacion <= 10:
                    calificaciones.append(calificacion)
                    break
                else:
                    print("⚠️ La calificación debe estar entre 0 y 10.")
            except ValueError:
                print("⚠️ Debe ingresar un número válido.")

        continuar = input("¿Desea ingresar otra materia? (s/n): ").lower()
        if continuar != "s":
            break

    return materias, calificaciones


def calcular_promedio(calificaciones):
    if len(calificaciones) == 0:
        return 0
    return sum(calificaciones) / len(calificaciones)


def determinar_estado(calificaciones, umbral=5.0):
    aprobadas = []
    reprobadas = []
    for i, nota in enumerate(calificaciones):
        if nota >= umbral:
            aprobadas.append(i)
        else:
            reprobadas.append(i)
    return aprobadas, reprobadas


def encontrar_extremos(calificaciones):
    if len(calificaciones) == 0:
        return None, None
    indice_max = calificaciones.index(max(calificaciones))
    indice_min = calificaciones.index(min(calificaciones))
    return indice_max, indice_min


# Función principal
if __name__ == "__main__":
    materias, calificaciones = ingresar_calificaciones()

    print("\n📊 Materias y calificaciones registradas:")
    for m, c in zip(materias, calificaciones):
        print(f"{m}: {c}")

    promedio = calcular_promedio(calificaciones)
    print(f"\n✅ Promedio general: {promedio:.2f}")

    aprobadas, reprobadas = determinar_estado(calificaciones)
    print("\n✅ Materias aprobadas:")
    for i in aprobadas:
        print(f"{materias[i]} ({calificaciones[i]})")

    print("\n❌ Materias reprobadas:")
    for i in reprobadas:
        print(f"{materias[i]} ({calificaciones[i]})")

    indice_max, indice_min = encontrar_extremos(calificaciones)
    print(f"\n📈 Materia con mayor calificación: {materias[indice_max]} ({calificaciones[indice_max]})")
    print(f"📉 Materia con menor calificación: {materias[indice_min]} ({calificaciones[indice_min]})")
```
